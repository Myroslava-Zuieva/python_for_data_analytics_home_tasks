1. З колоди у 52 карти випадково витягується одна карта. Яка ймовірність того, що це туз, якщо відомо, що ця карта — пікова?
2. У групі з 30 студентів є 18 хлопців і 12 дівчат. Відомо, що 7 хлопців і 5 дівчат носять окуляри. Випадково обраний студент носить окуляри. Яка ймовірність того, що цей студент — дівчина?
3. У конверті знаходяться 10 лотерейних білетів, серед яких 3 виграшних. З конверта послідовно витягуються білети. Знайти ймовірності того, що:
    
    а) 2-й витягнутий білет буде виграшним, якщо 1-й був виграшним;
    б) 3-й буде виграшним, якщо попередні два білети були виграшними;
    в) 4-й буде виграшним, якщо попередні білети були виграшними.
   

1. Ймовірність того, що карта — туз, якщо відомо, що ця карта пікова

В колоді 52 карти:

	•	Є 4 тузи (по одному в кожній масті),
	•	Є 13 пікових карт.

З цих 13 пікових карт лише 1 — туз піковий.

Ймовірність того, що вибрана карта — туз, за умови, що вона пікова, обчислюється за формулою умовної ймовірності:

$P(\text{туз} | \text{пікова}) = \frac{P(\text{туз піковий})}{P(\text{пікова})}$



Оскільки з 13 пікових карт лише 1 є тузом, то ймовірність дорівнює:

$P(\text{туз} | \text{пікова}) = \frac{1}{13}$


Отже, ймовірність того, що пікова карта — це туз, дорівнює  $\frac{1}{13} $.

2. Ймовірність того, що випадково обраний студент в окулярах — дівчина

Маємо:

	•	18 хлопців, з яких 7 носять окуляри.
	•	12 дівчат, з яких 5 носять окуляри.
	•	Всього 7 + 5 = 12 студентів носять окуляри.

Тепер використаємо формулу умовної ймовірності для обчислення ймовірності того, що студент в окулярах — дівчина:

$P(\text{дівчина} | \text{окуляри}) = \frac{P(\text{дівчина і окуляри})}{P(\text{окуляри})}$

де:

$P(\text{дівчина і окуляри}) = \frac{5}{30}$ — ймовірність того, що студент — дівчина і носить окуляри.
$P(\text{окуляри}) = \frac{12}{30}$  — ймовірність того, що студент носить окуляри.

Отже, ймовірність того, що студент в окулярах — дівчина:

$P(\text{дівчина} | \text{окуляри}) = \frac{\frac{5}{30}}{\frac{12}{30}} = \frac{5}{12}$


Отже, ймовірність того, що випадково обраний студент в окулярах — дівчина, дорівнює  $\frac{5}{12} $.

3. Ймовірність того, що:

а) 2-й витягнутий білет буде виграшним, якщо 1-й був виграшним

Всього є 10 білетів, з яких 3 — виграшні.

Якщо перший білет був виграшним, то залишається 9 білетів, з яких 2 — виграшні.

Ймовірність того, що другий білет виграшний:

$P(\text{2-й виграшний} | \text{1-й виграшний}) = \frac{2}{9}$


б) 3-й буде виграшним, якщо попередні два білети були виграшними

Якщо перші два білети були виграшними, то залишається 8 білетів, з яких тільки 1 — виграшний.

Ймовірність того, що третій білет виграшний:

$P(\text{3-й виграшний} | \text{2 попередні виграшні}) = \frac{1}{8}$


в) 4-й буде виграшним, якщо попередні білети були виграшними

Якщо перші три білети були виграшними, то залишається 7 білетів, і жоден з них не є виграшним. Отже, ймовірність того, що 4-й білет буде виграшним дорівнює 0.
