Тест дає позитивний результат у 99% випадків для хворих і 5% випадків для здорових людей (хибнопозитивний результат). Детектор брехні правильно ідентифікує чесних співробітників у 90% випадків і неправильно класифікує їх як таких, що брехали, у 10% випадків. Якщо детектор показав, що співробітник брехав, яка ймовірність того, що він насправді брехав?

Якщо предмет загублено в кімнаті, ймовірність знайти його під час пошуку складає 0,8 для першої кімнати, 0,6 для другої і 0,4 для третьої. Якщо випадково обрана продукція має дефект, яка ймовірність того, що вона була випущена третьою лінією?
5. Програма для розпізнавання рукописних цифр має точність 95% для цифри "3" і 98% для інших цифр. Якщо на вхід програми подано цифру "3", а програма її не розпізнала, яка ймовірність того, що насправді була подана цифра "3"? Відомо, що в 20% випадків на вхід подається цифра "3".

Задача 1

Спочатку я записала ймовірність того, що людина хвора:

$P(\text{хвора})$ = 0.01

	•	Потім ймовірність того, що тест дасть позитивний результат, якщо людина хвора:

$P(\text{позитивний результат} | \text{хвора})$ = 0.99

	•	Потім ймовірність того, що тест дасть позитивний результат, якщо людина здорова:

$P(\text{позитивний результат} | \text{здорова})$ = 0.05

	•	І в кінці ймовірність того, що людина здорова:

$P(\text{здорова}) = 1 - P(\text{хвора}) = 0.99$


Далі я шукала ймовірність того, що людина насправді хвора, якщо тест дав позитивний результат за допомогою формули Байєса.

$P(\text{хвора} | \text{позитивний результат}) = \frac{P(\text{позитивний результат} | \text{хвора}) \cdot P(\text{хвора})}{P(\text{позитивний результат})}$


Щоб знайти  $P(\text{позитивний результат})$ , використала теорему повної ймовірності:

$P(\text{позитивний результат}) = P(\text{позитивний результат} | \text{хвора}) \cdot P(\text{хвора}) + P(\text{позитивний результат} | \text{здорова}) \cdot P(\text{здорова})$

Підставила всі значення:

$P(\text{позитивний результат}) = (0.99 \cdot 0.01) + (0.05 \cdot 0.99) = 0.0099 + 0.0495 = 0.0594$


І обчислила ймовірність того, що людина насправді хвора, за умови позитивного результату тесту:

$P(\text{хвора} | \text{позитивний результат}) = \frac{0.99 \cdot 0.01}{0.0594} = \frac{0.0099}{0.0594} \approx 0.1667$


Отже, ймовірність того, що людина насправді хвора, якщо тест показав позитивний результат, дорівнює приблизно 0.167 або 16.67%.

Задача 2 

    •	Ймовірність того, що співробітник чесний: 
    
$P(\text{чесний}) = 0.70$

	•	Ймовірність того, що співробітник може брехати:

$P(\text{бреше}) = 0.30$

	•	Ймовірність того, що детектор правильно ідентифікує чесного співробітника:

$P(\text{детектор каже чесний} | \text{чесний}) = 0.90$

	•	Ймовірність того, що детектор неправильно класифікує чесного співробітника як такого, що брехав:

$P(\text{детектор каже брехав} | \text{чесний}) = 0.10$

	•	Позначила  p  як ймовірність того, що детектор правильно ідентифікує того, хто брехав:

$P(\text{детектор каже брехав} | \text{бреше}) = p$


Далі шукала ймовірність того, що співробітник дійсно брехав, якщо детектор показав, що він брехав, тобто  P(\text{бреше} | \text{детектор каже брехав}) за формулою Байєса


$P(\text{бреше} | \text{детектор каже брехав}) = \frac{P(\text{детектор каже брехав} | \text{бреше}) \cdot P(\text{бреше})}{P(\text{детектор каже брехав})}$


Щоб знайти  $P(\text{детектор каже брехав})$ , використала теорему повної ймовірності:

$P(\text{детектор каже брехав}) = P(\text{детектор каже брехав} | \text{бреше}) \cdot P(\text{бреше}) + P(\text{детектор каже брехав} | \text{чесний}) \cdot P(\text{чесний})$

Підставила значення:

$P(\text{детектор каже брехав}) = (p \cdot 0.30) + (0.10 \cdot 0.70) = 0.30p + 0.07$


$P(\text{бреше} | \text{детектор каже брехав}) = \frac{p \cdot 0.30}{0.30p + 0.07}$


Висновок:

Ймовірність того, що співробітник насправді брехав, якщо детектор показав, що він брехав, залежить від значення  p  — ймовірності того, що детектор правильно ідентифікує того, хто брехав.

Задача 3
Спочатку я визначила інформацію, що була нам дана:

	1. Ймовірність загубити предмет у кімнаті:
	•	Перша кімната:  P(A_1) = 0.5
	•	Друга кімната:  P(A_2) = 0.3
	•	Третя кімната:  P(A_3) = 0.2
	2.	Ймовірність знайти предмет, якщо він був загублений у кімнаті:
	•	Перша кімната:  P(B | A_1) = 0.8 
	•	Друга кімната:  P(B | A_2) = 0.6 
	•	Третя кімната:  P(B | A_3) = 0.4 


Далі шукала ймовірність того, що предмет було загублено в першій кімнаті за умови, що його знайшли, тобто  P(A_1 | B) за формулою Байєса:


$P(A_1 | B) = \frac{P(B | A_1) \cdot P(A_1)}{P(B)}$

Щоб знайти загальну ймовірність знайти предмет я мала порахувати  P(B)


$P(B) = P(B | A_1) \cdot P(A_1) + P(B | A_2) \cdot P(A_2) + P(B | A_3) \cdot P(A_3)$

Підставила значення:

$P(B) = 0.8 \cdot 0.5 + 0.6 \cdot 0.3 + 0.4 \cdot 0.2$


$P(B) = 0.4 + 0.18 + 0.08 = 0.66$


$P(A_1 | B) = \frac{0.8 \cdot 0.5}{0.66} = \frac{0.4}{0.66} \approx 0.606$


Ймовірність того, що предмет було загублено в першій кімнаті за умови, що його знайшли, становить приблизно 0.606 або 60.6%.

Задача 4
Спочатку я записала дані нам ймовірності:

    1. Ймовірність, що продукція була випущена певною лінією:
	•	Перша лінія:  P(A_1) = 0.5 
	•	Друга лінія:  P(A_2) = 0.3 
	•	Третя лінія:  P(A_3) = 0.2 
	2.	Ймовірність того, що продукція з певної лінії має дефект:
	•	Перша лінія:  P(B | A_1) = 0.02 
	•	Друга лінія:  P(B | A_2) = 0.03 
	•	Третя лінія:  P(B | A_3) = 0.05 


Далі я  шукала й мовірність того, що продукція була випущена третьою лінією за умови, що вона має дефект, тобто  P(A_3 | B) за формулою Байєса:


$P(A_3 | B) = \frac{P(B | A_3) \cdot P(A_3)}{P(B)}$

Далі обчислила  P(B)  — загальну ймовірність того, що продукція має дефект:


$P(B) = P(B | A_1) \cdot P(A_1) + P(B | A_2) \cdot P(A_2) + P(B | A_3) \cdot P(A_3)$

І підставила відомі нам значення:

$P(B) = 0.02 \cdot 0.5 + 0.03 \cdot 0.3 + 0.05 \cdot 0.2$


$P(B) = 0.01 + 0.009 + 0.01 = 0.029$


$P(A_3 | B) = \frac{0.05 \cdot 0.2}{0.029} = \frac{0.01}{0.029} \approx 0.345$


Ймовірність того, що продукція була випущена третьою лінією за умови, що вона має дефект, становить приблизно 0.345 або 34.5%.

Задача 5
Спочатку я визначила всі дані нам ймовірності:

    1.  Ймовірність того, що на вхід подано цифру “3”:

P(A_3) = 0.20

	2.	Ймовірність того, що на вхід подано будь-яку іншу цифру:

$P(A_{\text{not\_3}}) = 0.80$

	3.	Програма правильно розпізнає цифру “3” з точністю 95%:

$P(B_{\text{correct}} | A_3) = 0.95$

Тому ймовірність того, що програма не розпізнає цифру “3”, становить:

$P(B_{\text{wrong}} | A_3) = 1 - 0.95 = 0.05$

	4.	Програма правильно розпізнає інші цифри з точністю 98%, тобто ймовірність помилки для інших цифр:

$P(B_{\text{wrong}} | A_{\text{not\_3}}) = 1 - 0.98 = 0.02$



За допомогою теореми Байєса я шукала ймовірність того, що на вхід подали цифру “3”, за умови, що програма не розпізнала її, тобто  $P(A_3 | B_{\text{wrong}})$ .


$P(A_3 | B_{\text{wrong}}) = \frac{P(B_{\text{wrong}} | A_3) \cdot P(A_3)}{P(B_{\text{wrong}})}$


Спочатку обчислила  $P(B_{\text{wrong}})$  — загальну ймовірність того, що програма зробить помилку:

$P(B_{\text{wrong}}) = P(B_{\text{wrong}} | A_3) \cdot P(A_3) + P(B_{\text{wrong}} | A_{\text{not3}}) \cdot P(A{\text{not3}})$

А потім підставила значення:

$P(B{\text{wrong}}) = 0.05 \cdot 0.20 + 0.02 \cdot 0.80$


$P(B_{\text{wrong}}) = 0.01 + 0.016 = 0.026$


$P(A_3 | B_{\text{wrong}}) = \frac{0.05 \cdot 0.20}{0.026} = \frac{0.01}{0.026} \approx 0.3846$


Ймовірність того, що насправді була подана цифра “3”, за умови, що програма її не розпізнала, становить приблизно 0.3846 або 38.46%.
