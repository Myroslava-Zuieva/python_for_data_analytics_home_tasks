1. В урні знаходиться 15 білих, 5 червоних і 10 чорних куль. Випадковим чином витягується 1 куля, знайти ймовірність того, що вона буде: а) білою, б) червоною, в) чорною.
Загальна кількість куль:
15 білих + 5 червоних + 10 чорних = 30 куль.

Ймовірність витягнути білу кулю:

$P(\text{біла}) = \frac{\text{кількість білих куль}}{\text{загальна кількість куль}} = \frac{15}{30} = \frac{1}{2}$

Ймовірність витягнути червону кулю:

$P(\text{червона}) = \frac{\text{кількість червоних куль}}{\text{загальна кількість куль}} = \frac{5}{30} = \frac{1}{6}$

Ймовірність витягнути чорну кулю:

$P(\text{чорна}) = \frac{\text{кількість чорних куль}}{\text{загальна кількість куль}} = \frac{10}{30} = \frac{1}{3}$

1. В магазин надійшло 30 холодильників, п'ять з яких мають заводський дефект. Випадковим чином обирають один холодильник. Яка ймовірність того, що він буде без дефекту?

Загальна кількість холодильників у магазині — 30. Із них 5 мають заводський дефект, отже, холодильників без дефекту 30 - 5 = 25

Ймовірність того, що вибраний холодильник буде без дефекту, визначається як відношення кількості холодильників без дефекту до загальної кількості холодильників:

$P(\text{без дефекту}) = \frac{\text{кількість без дефекту}}{\text{загальна кількість}} = \frac{25}{30} = \frac{5}{6}$

Отже, ймовірність того, що вибраний холодильник буде без дефекту, дорівнює $\frac{5}{6}$.

1. Набираючи номер телефону, абонент забув дві останні цифри, але пам'ятає, що одна з них — нуль, а інша — непарна. Знайти ймовірність того, що він набере правильний номер.

Спочатку я визначили кількість можливих непарних цифр

Непарними є цифри: 1, 3, 5, 7, 9. Тому є 5 варіантів для вибору непарної цифри.

Далі визначила всі можливі варіанти двох останніх цифр

Оскільки одна цифра — 0, а інша — одна з непарних цифр, то можливі 2 варіанти розміщення цифр:

0 на передостанньому місці, непарна на останньому або непарна на передостанньому місці, 0 на останньому.

Таким чином, кількість усіх можливих варіантів набору номера становить:

2 * 5 = 10

Існує лише один правильний варіант серед цих 10 можливих варіантів. Тому ймовірність того, що абонент набере правильний номер:

$P(\text{правильний номер}) = \frac{1}{10}$

Отже, ймовірність правильного набору номера дорівнює  $\frac{1}{10} .$

1. Знайти ймовірність того, що при кидку двох гральних кубиків добуток чисел на них:
    
    а) дорівнюватиме семи; б) буде не меншим за 20; в) буде парним.
    
    На кожному кубику є 6 граней (цифри від 1 до 6), тому загальна кількість можливих комбінацій при кидку двох кубиків становить 6 * 6 = 36
    
    а) Ймовірність того, що добуток чисел дорівнюватиме семи
    
    Щоб добуток дорівнював семи є лише один варіант множення 1 * 7, на кубику немає числа 7, тому цей варіант неможливий (ймовірність = 0)
    
    б) Ймовірність того, що добуток чисел буде не меншим за 20
    
    Перерахуємо всі можливі комбінації, де добуток чисел дорівнює або більше 20:
    
    •	 4 * 5 = 20
    •	 5 * 4 = 20
    •	 4 * 6 = 24
    •	 6 * 4 = 24
    •	 5 * 5 = 25
    •	 5 * 6 = 30
    •	 6 * 5 = 30
    •	 6 * 6 = 36
    
    Отже, таких комбінацій є 8.
    
    Ймовірність:
    
    $P(\text{добуток} \geq 20) = \frac{8}{36} = \frac{2}{9}$
    
    в) Ймовірність того, що добуток чисел буде парним
    
    Спочатку я розглянула комбінації при яких обидва числа непарні (щоб добуток був теж непарним), в мене вийшов такий результат (1, 1), (1, 3), (1, 5), (3, 1), (3, 3), (3, 5), (5, 1), (5, 3), (5, 5)
    
    Тобто є 9 комбінацій, при яких добуток буде непарним.
    
    Відповідно, кількість комбінацій, де добуток парний:
    
    36 - 9 = 27
    
    Ймовірність того, що добуток парний:
    
    $P(\text{парний добуток}) = \frac{27}{36} = \frac{3}{4}$
    
2. Ймовірність того, що на тесті з біології учень О. правильно розв'яже більше 11 задач, дорівнює 0,67. Ймовірність того, що О. правильно розв'яже більше 10 задач, дорівнює 0,74. Знайдіть ймовірність того, що О. правильно розв'яже рівно 11 задач.

Спочатку я виписала інформацію про наявні в нас ймовірності:

•	Ймовірність того, що учень О. правильно розв’яже більше 11 задач:  $P({більше\ 11}) = 0,67 .$
•	Ймовірність того, що учень О. правильно розв’яже більше 10 задач:  $P({більше\ 10}) = 0,74 .$

Далі я виокремила ймовірність того, що О. правильно розв’яже більше 10 задач як суму двох подій:

1.	О. правильно розв’яже більше 11 задач.
2.	О. правильно розв’яже рівно 11 задач.

Тобто:

$P(\text{більше 10}) = P(\text{більше 11}) + P(\text{рівно 11})$

Тоді ймовірність того, що О. правильно розв’яже рівно 11 задач буде виглядати так

$P(\text{рівно 11}) = P(\text{більше 10}) - P(\text{більше 11}) = 0,74 - 0,67 = 0,07$
