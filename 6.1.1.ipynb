У першому завданні я рахувала за такою логікою, в першому місяці до нас приєдналося 100 підписників, другий місяць це (100-5%) + 150 = 245, третій місяць це (245 - 5%) + 170 = 402.75, можна округлити до 403.

Далі кожен місяць в нас був стабільний відтік 5%, тому я скористалася такою формулою 

$Sn = a* \frac{1 - r^n}{1 - r}$

Sn — загальна сума доходу за n місяців.
a — кількість підписників у першому місяці.
r — коефіцієнт зменшення (в нашому випадку r = 0.95).
n — кількість місяців.

$S22 = 403* \frac{1 - 0.95^22}{1-0.95}$

```abap
a = 403
r = 0.95
n = 22 
Sn = a * (1 - r**n) / (1 - r)
```

Цей код дав мені такий результат в 5452 користувача 

В кінці я додала результат за всі місяці (5452 + 100 + 245 + 403) * 10 доларів = 62 тисячі доларів

1. **Завдання 2**

Від січня 2020 року до грудня 2021 року кожен місяць в нас був приріст у 3% та відтік у 2%. 

N(n) = N(n-1) x (1 + приріст - відтік)

Тобто кількість активних користувачів у місяці = кількість активних користувачів у попередньому місяці помножити на 1 + приріст - відтік

Це буде N(n) = N(n-1) x 1.01

Початкове значення активних користувачів у січні 20020 року була 2000 користувачів

Далі я розрахувала кількість користувачів

Кроки:

1. **Початкове значення**: У січні 2020 року кількість активних користувачів була 2000.
2. **Щомісячне оновлення**: Кількість активних користувачів у кожному наступному місяці оновлюється з урахуванням щомісячного приросту і відтоку.

Тепер, розрахуємо кількість користувачів після 24 місяців (тобто стан на грудень 2021 року):

### N(24) = 2000 x (1.01) в 24 степені = 2000 х 1.2682 = 2536.4

Тому, на кінець грудня 2021 року кількість активних користувачів буде приблизно **2536**.

1. **Завдання 3**

Спочатку я окремо порахувала кількість користувачів за перші три місяці:

1.	Січень 2021 року:
•	У січні залучено 500 нових користувачів.
2.	Лютий 2021 року:
•	Приріст користувачів на 50% від січневого числа.
•	Кількість нових користувачів у лютому = 500 + 50% = 500 × 1.5 = 750.
3.	Березень 2021 року:
•	Приріст користувачів на 20% від лютневого числа.
•	Кількість нових користувачів у березні = 750 + 20% = 750 × 1.2 = 900.

Далі, коли приріст стабілізувався до 10% на місяць, я рахувала за допомогою формули:

Sn = a * rn - 1 / r - 1

Де Sn  - сума перших  n  членів прогресії (загальна кількість користувачів за 3 місяці)

a  — кількість нових користувачів у першому місяці (у нашому випадку квітень,  a = 900 )

r  — коефіцієнт приросту (у нашому випадку  r = 1.1 , що відповідає приросту 10%)

 n  — кількість місяців (тут  n = 3 )

$S3 = 900 * \frac{1.1^3 - 1}{1.1-1}$

$S3 = 900 * \frac{1.331 - 1}{0.1} = 900 * \frac{0.331}{0.1} = 900*3.31 = 2979$

І в кінці треба додати всіх користувачів 500 + 750 + 900 + 2979 = 5129 користувачів
